```python
from sklearn.linear_model import LinearRegression
import numpy as np

# Assuming you have your input features X and target variable y
X1_engine_size_cylinders = np.array([[2.0,4],[2.4,4],[1.5,4],[3.5,6],[3.5,4],[3.5,6],[3.7,5],[3.7,5]])

y = np.array([196,221,136,255,244,275,275,275])

# Create a linear regression model
model = LinearRegression()

# Fit the model to the data
n=model.fit(X1_engine_size_cylinders, y)

# Predict the target variable for new input features
new_X1 = np.array([[2.4,4],[3.5,6],[3.7,5],[4.0,8],[2.1,3]])

predictions = model.predict(new_X1)

print(f'predictions are {predictions}')
import matplotlib.pyplot as plt
# Plot the consecutive predictions
plt.plot(predictions)
plt.xlabel('Index')
plt.ylabel('Prediction')
plt.title('Consecutive Predictions')
plt.show()
plt.scatter(range(len(y)), y, color='red', label='True Values')
plt.legend()
plt.show()
```
